```python
import os
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.autograd import Variable as V
import torchvision
import torchvision.transforms as transforms
import torch.optim as optim

import matplotlib.pyplot as plt
import numpy as np

import pyro
from pyro.distributions import Normal
from pyro.infer import SVI, Trace_ELBO
from pyro.optim import Adam
# for CI testing
smoke_test = ('CI' in os.environ)
pyro.enable_validation(True)
```

```python
transform = transforms.Compose(
    [
        transforms.ToTensor(),
        lambda x : x/x.max()
#         transforms.Normalize((0.1307,), (0.3081,)) #normalization values for mnist
    ]
)

batch_size=4

trainset = torchvision.datasets.MNIST(root='./data', train=True,
                                        download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=batch_size,
                                          shuffle=True, num_workers=2)

testset = torchvision.datasets.MNIST(root='./data', train=False,
                                       download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=batch_size,
                                         shuffle=False, num_workers=2)
```

```python
class Encoder(nn.Module):
    '''
    Expected a 1X28X28 image patch
    output will be 3X7X7
    '''
    def __init__(self):
        super(Encoder, self).__init__()
        self.conv1 = nn.Conv2d(1, 64, 1, stride=2, padding=0)
        self.conv2 = nn.Conv2d(64, 64, 3, stride=2, padding=1)
        self.conv3 = nn.Conv2d(64, 3, 1, stride=1, padding=0)
    
    def forward(self, x):
        y = F.relu(self.conv1(x)) #64,128,128
        y = F.relu(self.conv2(y)) #64,64,64
        y = F.relu(self.conv3(y))
        return y
    
class Decoder(nn.Module):
    '''
    Expected a 3X7X7 image patch
    output will be 1X28X28
    '''
    def __init__(self):
        super(Decoder, self).__init__()
        self.conv1 = nn.ConvTranspose2d(3, 64, 2, stride=2, padding=0)
        self.bn1 = nn.BatchNorm2d(64)
        self.conv2 = nn.ConvTranspose2d(64, 64, 2, stride=2, padding=0)
        self.bn2 = nn.BatchNorm2d(64)
        self.conv3 = nn.Conv2d(64, 32, 1, stride=1, padding=0)
        self.bn3 = nn.BatchNorm2d(32)
        self.conv4 = nn.Conv2d(32, 1, 3, stride=1, padding=1)
        self.bn4 = nn.BatchNorm2d(1)
     
    def forward(self, x):
        y = F.relu(self.bn1(self.conv1(x)))
        y = F.relu(self.bn2(self.conv2(y))) 
        y = F.relu(self.bn3(self.conv3(y)))
        y = F.sigmoid(self.bn4(self.conv4(y))) 
        return y

class ED(nn.Module):
    '''
    Expected E,D
    '''
    def __init__(self):
        super(ED, self).__init__()
        self.E = nn.ModuleList([Encoder()])
        self.D = nn.ModuleList([Decoder()])
    def forward(self, x):
        h_enc = self.E[0](x)
#         print('encoder out checking for nan ', np.isnan(h_enc.data.cpu()).any())
        y = self.D[0](h_enc)
#         print('decoder out checking for nan ', np.isnan(y.data.cpu()).any())
        return y
```

## No uncertainty

```python
log_interval = 10000

def train(model, train_loader, optimizer, epoch):
    model.train()
    for batch_idx, (data, target) in enumerate(train_loader):
        data, target = V(data.cuda()), V(target.cuda())
        optimizer.zero_grad()
        output = model(data)
        loss = F.mse_loss(output, data) #autoencoder
        loss.backward()
        optimizer.step()
        if batch_idx % log_interval == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx*len(data), len(train_loader.dataset),
                100.*batch_idx/len(train_loader), loss.data.item()))

def test(model, test_loader):
    model.eval()
    test_loss = 0
    for data, target in test_loader:
        data, target = V(data.cuda(), requires_grad=False), V(target.cuda(), requires_grad=False)
        output = model(data)
        test_loss += F.mse_loss(output, data).data.item() # sum up batch loss
        output_npy = output.data.cpu().numpy()
        for i in range(output_npy.shape[0]):
            plt.figure(figsize=(15,5))
            plt.subplot(1,2,1)
            plt.imshow(data.data.cpu().numpy()[i,0,:,:], 'gray')
            plt.title('input')
            
            plt.subplot(1,2,2)
            plt.imshow(output_npy[i,0,:,:], cmap='gray')
            plt.title('reconstructed')
            plt.show()
        break

    test_loss /= len(test_loader.dataset)
    print('\nTest set: Average loss: {:.4f}\n'.format(test_loss))
```

```python
model = ED()
model = model.cuda()
optimizer = optim.Adam(model.parameters(), lr=0.0001, weight_decay=0)
for eph in range(30):
    print('eph {}...'.format(eph))
    train(model, trainloader, optimizer, eph)
    if eph%10 == 0:
        test(model, testloader)
```

## Modelling Uncertainty

```python
Net = ED()
Net = Net.cuda()
```

```python
print(Net.D[0].conv3.weight.size(), Net.D[0].conv3.bias.size()) 
```

```python
def model(data):
    ##for conv layer
    loc1, scale1 = torch.zeros(32, 64, 1, 1), 10*torch.ones(32, 64, 1, 1)
    bias_loc1, bias_scale1 = torch.zeros(32), 10*torch.ones(32)
    w_prior1 = Normal(loc1, scale1).independent(4)
#     print('w_prior1 shapes: ', w_prior1.batch_shape, w_prior1.event_shape)
    b_prior1 = Normal(bias_loc1, bias_scale1).independent(1)
#     print('b_prior1 shapes: ', b_prior1.batch_shape, b_prior1.event_shape)
    
    priors = {'conv3.weight': w_prior1, 'conv3.bias': b_prior1}
    # lift module parameters to random variables sampled from the priors
    lifted_module = pyro.random_module("module", Net, priors)
    # sample a regressor (which also samples w and b)
    lifted_reg_model = lifted_module()
    with pyro.iarange("map", batch_size):
        x_data = data[0]
        y_data = data[1]
        # run the regressor forward conditioned on data
        prediction_mean = lifted_reg_model(x_data)
#         print('pred_mean shape: ', prediction_mean.size())
#         print('sample shape : ',Normal(prediction_mean, 0.1*torch.ones(data.size(0), 1, outdim)).independent(2).sample().shape)
        # condition on the observed data
        pyro.sample("obs",
                    Normal(prediction_mean, 0.1*torch.ones(batch_size, 1, 28, 28).cuda()).independent(3),
                    obs=y_data)
```

```python
softplus = torch.nn.Softplus()

def guide(data):
    # define our variational parameters
    
    ##for conv layer
    w_loc1 = torch.randn(32, 64, 1, 1)
    w_log_sig1 = torch.tensor(-3.0*torch.ones(32, 64, 1, 1) + 0.05*torch.randn(32, 64, 1, 1))
    b_loc1 = torch.randn(32)
    b_log_sig1 = torch.tensor(-3.0*torch.ones(32) + 0.05*torch.randn(32))
  
    # register learnable params in the param store    
    mw_param1 = pyro.param("guide_mean_weight1", w_loc1)
    sw_param1 = softplus(pyro.param("guide_log_scale_weight1", w_log_sig1))
    mb_param1 = pyro.param("guide_mean_bias1", b_loc1)
    sb_param1 = softplus(pyro.param("guide_log_scale_bias1", b_log_sig1))
    # guide distributions for w and b
    w_dist1 = Normal(mw_param1, sw_param1).independent(4)
    b_dist1 = Normal(mb_param1, sb_param1).independent(1)
    dists = {'conv3.weight':w_dist1, 'conv3.bias':b_dist1}
    # overload the parameters in the module with random samples
    # from the guide distributions
    lifted_module = pyro.random_module("module", Net, dists)
    # sample a regressor (which also samples w and b)
    return lifted_module()

optim = Adam({"lr": 0.05})
svi = SVI(model, guide, optim, loss=Trace_ELBO())
```

```python
num_iterations = 1 if not smoke_test else 2
def main():
    pyro.clear_param_store()
    for j in range(num_iterations):
        print(j, '...')
        # calculate the loss and take a gradient step
        for batch_idx, XY in enumerate(trainloader):
#             print(XY[0].size(), XY[1].size())
            x_data = V(XY[0].cuda())
            loss = svi.step([x_data, x_data])
            if batch_idx%100 == 0:
                print("[iteration {}] loss: {:.4f}".format(batch_idx, loss/float(batch_size)))

if __name__ == '__main__':
    main()
```

```python
for batch_idx, XY in enumerate(testloader):
    y_preds = []
    x_data = V(XY[0].cuda())
    for i in range(20):
        sampled_reg_model = guide(None)
        # run the regression model and add prediction to total
        y_preds.append(sampled_reg_model(x_data).data.cpu().numpy())
    y_preds = np.array(y_preds)
    # take the average of the predictions, std will represent uncertainty per pixel
    mean_image = np.mean(y_preds, axis=0)
    std_image = np.std(y_preds, axis=0)
    output_npy = mean_image
    for i in range(output_npy.shape[0]):
        plt.figure(figsize=(15,5))
        plt.subplot(1,3,1)
        plt.imshow(x_data.data.cpu().numpy()[i,0,:,:], 'gray')
        plt.title('input')

        plt.subplot(1,3,2)
        plt.imshow(output_npy[i,0,:,:], cmap='gray')
        plt.title('reconstructed')
        
        plt.subplot(1,3,3)
        plt.imshow(std_image[i,0,:,:], cmap='gray')
        plt.title('std')
        plt.show()
    if batch_idx == 2:
        break
```
